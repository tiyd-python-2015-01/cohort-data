```python
import pandas as pd
import numpy as md
import datetime
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rc('figure', figsize=(16, 6))
```

```python
python='cohort_3_python.csv'
pythondf = pd.read_csv(python, skipfooter=4, engine='python')
pythondf.head(2)
```

```python
pythondf = pd.melt(pythondf, id_vars=['Name'], var_name='Title_Date', value_name='Score')
pythondf.head(2)
```

```python
pythondf['Course'] = "Python"
pythondf.head(2)
```

```python
def reformat_date(string_date):
    return datetime.datetime.strptime(string_date+"2015 07", '%b%d%Y %H')

def reformat_python_date(pythondf):
    for i, myfield in pythondf.iterrows(): 
        date = myfield['Title_Date'].split(',')
        if len(date) == 2:
            if date[1].find("21") != -1:
                date[1] = 'Jan21'   
            pythondf.loc[i,'Date'] = reformat_date(date[1].replace(" ",""))
        else:
            if date[0].find("Charting") != -1:
                pythondf.loc[i,'Date'] = reformat_date("Jan28")
            elif date[0].find("PigSim") != -1:
                pythondf.loc[i,'Date'] = reformat_date("Jan29")
            elif date[0].find("Traffic") != -1:
                pythondf.loc[i,'Date'] = reformat_date("Feb2") 
        if myfield['Title_Date'].find("Lecture") == -1:
           pythondf.loc[i, 'Type'] = "HW"
        else:
            pythondf.loc[i,'Type'] = "Lecture"

reformat_python_date(pythondf)   
```

```python
del(pythondf['Title_Date'])
pythondf.head(20)
```

Now we work with the Ruby Excel worksheets

```python
ruby='cohort_3_rails.xlsx'
```

```python
ruby_lecture = pd.read_excel(ruby,
                             'Lecture Score', 
                             index_col=None,
                             na_values=['NA'],
                             skiprows=1,
                             skip_footer=3,
                             header=0,
                             parse_cols='A:N')

ruby_lecture.head(2)
```

```python
ruby_hw = pd.read_excel(ruby,
                        'HW Score',
                        index_col=None,
                        na_values=['NA'],
                        skiprows=1,
                        skip_footer=2,
                        header=0,
                        parse_cols='A:N')
ruby_hw.head(2)
```

```python
ruby_lecturedf = pd.melt(ruby_lecture, id_vars='Unnamed: 0', var_name='Date', value_name='Score')
ruby_lecturedf['Type'] = "Lecture"
ruby_lecturedf.head(2)
```

```python
ruby_hwdf = pd.melt(ruby_hw, id_vars='Unnamed: 0', var_name='Date', value_name='Score')
ruby_hwdf['Type'] = "HW"
ruby_hwdf.head(2)
```

```python
ruby_all = pd.concat([ruby_hwdf,ruby_lecturedf])
ruby_all.tail(10)
```

```python
ruby_all.head(10)
```

```python
ruby_all.columns.values[0]='Name'
ruby_all['Course'] = "Ruby"
ruby_all.head(2)
```

```python
def process_ruby_date(rubydf):
    for i, myfield in rubydf.iterrows(): 
        date1 = myfield['Date'].replace(" ","")
        if date1 == "M":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(12))
        elif date1 =="T":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(13))
        elif date1 =="W":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(14))
        elif date1 =="Th":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(15))
        elif date1 == "M.1":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(19))
        elif date1 =="T.1":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(20))
        elif date1 =="W.1":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(21))
        elif date1 =="Th.1":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(22))
        elif date1 == "M.2":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(26))
        elif date1 =="T.2":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(27))
        elif date1 =="W.2":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(28))
        elif date1 =="Th.2":
            rubydf.loc[i,'Date'] = reformat_date("Jan"+str(29))  
        if date1 == "M.3":
            rubydf.loc[i,'Date'] = reformat_date("Feb"+str(2))
        
    
process_ruby_date(ruby_all)
ruby_all.head(20)
```

```python
cohortdf = pd.concat([ruby_all,pythondf])
```

```python
lecturedf = cohortdf[cohortdf['Type'] == "Lecture"].dropna()
lecturedf_ruby = lecturedf[lecturedf['Course'] == "Ruby"]
del lecturedf_ruby['Course']
lecturedf_python = lecturedf[lecturedf['Course'] == "Python"]
del lecturedf_python['Course']

hwdf = cohortdf[cohortdf['Type'] == "HW"].dropna()
hwdf_ruby = hwdf[hwdf['Course'] == "Ruby"]
del hwdf_ruby['Course']
hwdf_python = hwdf[hwdf['Course'] == "Python"]
del hwdf_python['Course']
```

```python
lecturedf_ruby.set_index(['Date'])
lecturedf_python.set_index(['Date'])
lecturedf_ruby.index = pd.to_datetime(lecturedf_ruby['Date'], format="%Y%m%d")
lecturedf_python.index = pd.to_datetime(lecturedf_python['Date'], format="%Y%m%d")
lecturedf_ruby['day'] = lecturedf_ruby.index.day
lecturedf_python['day'] = lecturedf_python.index.day

lec_ruby_mean = lecturedf_ruby.groupby(['day']).mean()
lec_python_mean = lecturedf_python.groupby(['day']).mean()
```

```python
plt.xticks(range(len(lec_ruby_mean)), lec_ruby_mean.index)
plt.plot(lec_ruby_mean,label = "Ruby")
plt.plot(lec_python_mean, label = "Python")
plt.legend(loc='upper left')
plt.title("Mean Daily Scores for Lectures by Cohort")
plt.show()
```
