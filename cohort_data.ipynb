```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import re
%matplotlib inline
```

```python
python_lecture_data = pd.read_excel("cohort_3_python_lectures.xlsx")
python_homework_data = pd.read_excel("cohort_3_python_homework.xlsx")
```

```python
rails_lecture_data = pd.read_excel("cohort_3_rails_clean.xlsx")
rails_homework_data = pd.read_excel("cohort_3_rails_clean.xlsx", sheetname=1)
```

```python
python_lecture_data_transposed = python_lecture_data.transpose()
python_lecture_data_transposed.index = pd.to_datetime(python_lecture_data_transposed.index, format="%m/%d/%y")
```

```python
python_homework_data_transposed = python_homework_data.transpose()
python_homework_data_transposed.index = pd.to_datetime(python_homework_data_transposed.index, format="%m/%d/%y")
```

```python
rails_lecture_data_transposed = rails_lecture_data.transpose()
rails_lecture_data_transposed.index = pd.to_datetime(rails_lecture_data_transposed.index, format="%m/%d/%y")
```

```python
rails_homework_data_transposed = rails_homework_data.transpose()
rails_homework_data_transposed.index = pd.to_datetime(rails_homework_data_transposed.index, format="%m/%d/%y")
```

```python
py_lecture_by_day = python_lecture_data_transposed.mean(1)
labels = ["Lecture {}".format(x) for x in range(1, 14)]
py_lecture_by_day.plot(kind="bar", title="Mean Difficulty of Python Lectures by Day")
plt.xticks(np.arange(13), labels)
plt.show()
```

From the above graph we can see that as the course has progressed the average difficulty rating by students has increased, especially lectures toward the end of the week and the most recent lectures.

```python
py_homework_by_day = python_homework_data_transposed.mean(1)
labels = ["Homework {}".format(x) for x in range(1, 12)]
py_homework_by_day.plot(kind="bar", title="Mean Difficulty of Python Homework by Day")
plt.xticks(np.arange(11), labels)
plt.show()
```

The difficulty of the Python homework assignements have gone steadily up since the beginning of the course.

```python
rails_lecture_by_day = rails_lecture_data_transposed.mean(1)
labels = ["Lecture {}".format(x) for x in range(1, 14)]
rails_lecture_by_day.plot(kind="bar", title="Mean Difficulty of Rails Lectures by Day")
plt.xticks(np.arange(13), labels)
plt.show()
```

The rails lectures have stayed consistantly difficult, save the first two days of the course and one day on Jan 23.

```python
rails_homework_by_day = rails_homework_data_transposed.mean(1)
labels = ["Homework {}".format(x) for x in range(1, 14)]
rails_homework_by_day.plot(kind="bar", title="Mean Difficulty of Rails Homework by Day")
plt.xticks(np.arange(13), labels)
plt.show()
```

Rails homeworks seems to be consistantly difficult, with the most difficult assignment occuring on Jan 21.

```python
python_lecture_week_1 = python_lecture_data_transposed[0:4].mean().mean()
python_lecture_week_2 = python_lecture_data_transposed[4:8].mean().mean()
python_lecture_week_3 = python_lecture_data_transposed[8:12].mean().mean()
python_lecture_week_4 = python_lecture_data_transposed[12:].mean().mean()
```

```python
labels = ["Week {}".format(x) for x in range(1,5)]
pd.DataFrame([python_lecture_week_1, python_lecture_week_2, python_lecture_week_3, python_lecture_week_4]).plot(kind="bar", 
                                                                                title="Python Lecture Difficulty By Week")
plt.xticks(np.arange(0,4), labels)
plt.show()
```

Python lecture difficulties have stayed fairly consistant over the past couple of weeks.  Week 4 data is not really accurate because it contains only a single data point.

```python
python_homework_week_1 = python_homework_data_transposed[0:4].mean().mean()
python_homework_week_2 = python_homework_data_transposed[4:8].mean().mean()
python_homework_week_3 = python_homework_data_transposed[8:12].mean().mean()
python_homework_week_4 = python_homework_data_transposed[12:].mean().mean()
```

```python
pd.DataFrame([python_homework_week_1, python_homework_week_2, python_homework_week_3, python_homework_week_4]).plot(kind="bar", 
                                                                                title="Python homework Difficulty By Week")
plt.xticks(np.arange(0,4), labels)
plt.show()
```

Python homeworks have become increasingly difficult as the course has progressed.

```python
rails_lecture_week_1 = rails_lecture_data_transposed[0:4].mean().mean()
rails_lecture_week_2 = rails_lecture_data_transposed[4:8].mean().mean()
rails_lecture_week_3 = rails_lecture_data_transposed[8:12].mean().mean()
rails_lecture_week_4 = rails_lecture_data_transposed[12:].mean().mean()
```

```python
pd.DataFrame([rails_lecture_week_1, rails_lecture_week_2, rails_lecture_week_3, rails_lecture_week_4]).plot(kind="bar", 
                                                                                title="Rails Lecture Difficulty By Week")
plt.xticks(np.arange(0,4), labels)
plt.show()
```

Rails lectures have had a near linear increase in difficulty over the first four weeks.

```python
pd.DataFrame([rails_homework_week_1, rails_homework_week_2, rails_homework_week_3, rails_homework_week_4]).plot(kind="bar", 
                                                                                title="Rails Homework Difficulty By Week")
plt.xticks(np.arange(0,4), labels)
plt.show()
```

The difficulty of the homework assignments increases almost linearly throughout the week, with the most difficult assignments being the weekend assignments.
