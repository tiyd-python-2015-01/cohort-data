### Mean difficulty of lectures per week per class:

```python
ruby_lecture_means
```

```python
week1_means = ruby_lecture_means[['M1', 'T1', 'W1', 'Th1']].mean()
week2_means = ruby_lecture_means[['T2', 'W2', 'Th2']].mean()
week3_means = ruby_lecture_means[['M3', 'T3', 'W3', 'Th3']].mean()
week4_means = ruby_lecture_means['M4'].mean()
weekly_means = [week1_means, week2_means, week3_means, week4_means]
print("Ruby Weekly Means:")
for _ in range(4):
    print("Week {}: {}".format((_+1), weekly_means[_]))

plt.plot(weekly_means)
plt.xticks(range(4), ["Week 1", "Week 2", "Week 3", "Week 4"])
plt.title("Graph of Ruby Lecture Means by Week")
plt.ylabel("Difficulty Rating")
plt.xlabel("Week")
plt.ylim(ymin=1, ymax=6)
plt.show()
```
