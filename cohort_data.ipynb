```python
import pandas as pd
import matplotlib.pyplot as ply
import seaborn as sbn
import numpy as np

%matplotlib inline
```

```python
cohort_rails = pd.ExcelFile ("cohort_3_rails.xlsx")
ru_lec = cohort_rails.parse('Lecture Score')
ru_hw = cohort_rails.parse('HW Score')

ru_lec.drop(['Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Week 5', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Week 6', 'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Week 7', 'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27', 'Week 8', 'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31', 'Week 9', 'Unnamed: 33', 'Unnamed: 34', 'Unnamed: 35'],inplace=True,axis=1)
ru_hw.drop(['Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Week 5', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Week 6', 'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Week 7', 'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27', 'Week 8', 'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31', 'Week 9', 'Unnamed: 33', 'Unnamed: 34', 'Unnamed: 35'],inplace=True,axis=1)
ru_lec.columns = ['lec1', 'lec2', 'lec3', 'lec4', 'lec5', 'lec6', 'lec7', 'lec8', 'lec9', 'lec10', 'lec11', 'lec12', 'lec13']
ru_hw.columns = ['hw1', 'hw2', 'hw3', 'hw4', 'hw5', 'hw6', 'hw7', 'hw8', 'hw9', 'hw10', 'hw11', 'hw12', 'hw13']

ru_combined = ru_lec.join(ru_hw)
ru_combined = ru_combined[pd.notnull(ru_combined['hw9'])]
ru_combined = ru_combined.ix[1:-2]
ru_combined = ru_combined.dropna(axis=1,how='all')
```

```python
cohort_python = pd.read_csv("cohort_3_python.csv")
cohort_python.columns = ['Name', 'lec1', 'hw1', 'lec2', 'hw2', 'lec3', 'hw3', 'lec4', 'hw4', 'lec5', 'hw5', 'lec6', 'hw6', 'lec7', 'lec8', 'hw8', 'lec9', 'hw9', 'lec10', 'hw10', 'lec11', 'hw11', 'lec12', 'hw12', 'lec13']
cohort_python.drop(['lec13'],inplace=True,axis=1)
cohort_python = cohort_python.dropna(axis=1,how='all')
cohort_python = cohort_python.dropna(axis=0,how='all')
```

# Deliverables

###### Mean difficulty for lectures per day, per class

```python
py_lectures = [col for col in cohort_python.columns if 'lec' in col]
py_homework = [col for col in cohort_python.columns if 'hw' in col]

ru_lectures = [col for col in ru_combined.columns if 'lec' in col]
ru_homework = [col for col in ru_combined.columns if 'hw' in col]
```

```python
py_lec_mean = cohort_python[py_lectures][:14].mean(axis=0)
ru_lec_mean = ru_combined[ru_lectures][:14].mean(axis=0)
```

```python
ply.plot(py_lec_mean, c='red', label='Python')
ply.plot(ru_lec_mean, c='blue', label='Ruby')
ply.xlabel('day')
ply.ylabel('mean scoring per student')
ply.title('Mean difficulty for lectures per day, per class')
ply.legend(loc=4)
ply.show()
```

###### Choose a sample student from each class and graph their data. What's their story? Are they getting better over time?

```python
ruby_student = ru_lec[6:7]
python_student = cohort_python[7:8]

py_lectures = [col for col in cohort_python.columns if 'lec' in col]
rub_lect_col = [col for col in ru_lec.columns if 'lec' in col]
```

```python
python_student[py_lectures].max().plot(c='red', label='Python Student')
ruby_student[rub_lect_col].max().plot(c='blue', label='Ruby Student')
ply.xlabel('Day')
ply.ylabel('mean lecture scoring per student')
ply.title('Difficulty for 1 student per class')
ply.legend(loc=4)
ply.show()
```

The ruby student is maintaining an average pace in regards to how they view lectures. The python student on the other hand is progressively getting more and more challenged in lectures every day. They have a steady growth into pain. As seen in week1/week2, the Ruby had one rough day, but otherwise has hovered between 4 and 4.5.
