```python
import pandas as pd
import matplotlib.pyplot as ply
import seaborn as sbn

%matplotlib inline
```

```python
cohort_python = pd.read_csv("cohort_3_python.csv")
cohort_python.head()
```

```python
cohort_rails = pd.ExcelFile ("cohort_3_rails.xlsx")
ru_lec = cohort_rails.parse('Lecture Score')
ru_hw = cohort_rails.parse('HW Score')
```

# Deliverables

###### Mean difficulty for lectures per day, per class

```python
py_lectures = [col for col in cohort_python.columns if 'Lecture' in col]
py_homework = [col for col in cohort_python.columns if 'Lecture' not in col]

rub_lect_col = ru_lec.columns[:13]
rub_hw_col = ru_hw.columns[:13]
```

```python
py_lec_mean = cohort_python[py_lectures][:16].mean(axis=0)
ru_lec_mean = ru_lec[rub_lect_col][1:15].mean(axis=0)
```

```python
ply.plot(py_lec_mean, c='red', label='Python')
ply.plot(ru_lec_mean, c='blue', label='Ruby')
ply.xlabel('day')
ply.ylabel('mean scoring per student')
ply.title('Mean difficulty for lectures per day, per class')
ply.legend(loc=4)
ply.show()
```

###### Mean difficulty for lectures per week, per class

```python
py_lec_week1_mean = py_lec_mean[0:4].mean(axis=0)
py_lec_week2_mean = py_lec_mean[4:8].mean(axis=0)
py_lec_week3_mean = py_lec_mean[8:12].mean(axis=0)
py_lec_week4_mean = py_lec_mean[12:].mean(axis=0)
py_lec_weekly_mean = [py_lec_week1_mean, py_lec_week2_mean, py_lec_week3_mean, py_lec_week4_mean]

ru_lec_week1_mean = ru_lec_mean[0:4].mean(axis=0)
ru_lec_week2_mean = ru_lec_mean[4:8].mean(axis=0)
ru_lec_week3_mean = ru_lec_mean[8:12].mean(axis=0)
ru_lec_week4_mean = ru_lec_mean[12:].mean(axis=0)
ru_lec_weekly_mean = [ru_lec_week1_mean, ru_lec_week2_mean, ru_lec_week3_mean, ru_lec_week4_mean]
```

```python
ply.plot(py_lec_weekly_mean, c='red', label='Python')
ply.plot(ru_lec_weekly_mean, c='blue', label='Ruby')
ply.xlabel('Week')
ply.ylabel('mean scoring per student')
ply.title('Mean difficulty for lectures per week, per class')
ply.legend(loc=4)
ply.show()
```

###### Choose a sample student from each class and graph their data. What's their story? Are they getting better over time?

```python
ruby_student = ru_lec[6:7]
python_student = cohort_python[7:8]

py_columns = [column for 
              column in 
              python_student[py_lectures].columns 
              if python_student[column].dtype == 'float64']
```

```python
python_student[py_columns].max().plot(c='red', label='Python Student')
ruby_student[rub_lect_col].max().plot(c='blue', label='Ruby Student')
ply.xlabel('Day')
ply.ylabel('mean lecture scoring per student')
ply.title('Difficulty for 1 student per class')
ply.legend(loc=4)
ply.show()
```

The ruby student is maintaining an average pace in regards to how they view lectures. The python student on the other hand is progressively getting more and more challenged in lectures every day. They have a steady growth into pain. As seen in week1/week2, the Ruby had one rough day, but otherwise has hovered between 4 and 4.5.
