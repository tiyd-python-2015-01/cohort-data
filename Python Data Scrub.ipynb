# Lecture Stuff

```python
# piece1 = python[0:14:2]
# piece2 = python[15::2]
# lecture = pd.concat(piece1, piece2)

#broke out all the Lecture rows and them put them back together
pieces = [python[0:13:2], python[13::2]]
lecture = pd.concat(pieces)
lecture
```

```python
lect_mean = lecture[::].mean(axis=1)
lect_mean
```

```python
lect_mean.plot(kind='bar', fontsize=20, figsize=(15, 7))
```

```python
#python
weeks = [lecture[0:4], lecture[4:8], lecture[8:12]]
lecture_weeks = pd.concat(weeks, keys=['Week1', 'Week2', 'Week3'])
lecture_weeks
```

```python
lecture_weeks.ix['Week1']
```

```python
#week1_mean = lecture_weeks.ix['Week1'].mean()
#axis=0 in the mean function
week1_mean = (lecture_weeks.ix['Week1'].mean()).mean()
week2_mean = (lecture_weeks.ix['Week2'].mean()).mean()
week3_mean = (lecture_weeks.ix['Week3'].mean()).mean()
print(week1_mean)
print(week2_mean)
print(week3_mean)
```

```python
#test_mean = lecture_weeks.ix["Week1", "Week2", "Week3"].mean()
#***THIS DOESN'T WORK :( ..... KeyError: 'Key length (3) exceeds index depth (2)'
#test_mean
```

```python
weekly_mean = [week1_mean, week2_mean, week3_mean]
labels = ['Week1', 'Week2', 'Week3']
week_mean = pd.Series(weekly_mean, index=labels)
week_mean.plot(kind='bar', fontsize=20, figsize=(10, 5))
```

```python
hwsplit = [python[1:12:2], python[14::2]]
homework = pd.concat(hwsplit)
homework
```

```python
hw_mean = homework[::].mean(axis=1)
hw_mean
```

```python
hw_mean.plot(kind='bar', fontsize=20, figsize=(15, 7))
```

```python
hw_weeks = [homework[0:3], homework[3:6], homework[6:11]]
hw_weeks = pd.concat(hw_weeks, keys=['Week1', 'Week2', 'Week3'])
hw_weeks
```

```python
hw_week1_mean = (hw_weeks.ix['Week1'].mean()).mean()
hw_week2_mean = (hw_weeks.ix['Week2'].mean()).mean()
hw_week3_mean = (hw_weeks.ix['Week3'].mean()).mean()
print(hw_week1_mean)
print(hw_week2_mean)
print(hw_week3_mean)
```

```python
hwweek_mean = [hw_week1_mean, hw_week2_mean, hw_week3_mean]
labels = ['Week1', 'Week2', 'Week3']
hwweek_mean = pd.Series(hwweek_mean, index=labels)
hwweek_mean.plot(kind='bar', fontsize=20, figsize=(10, 5))
```
